```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
# Functional methode 

plt.plot(x,y,'r-')
plt.xlabel('X lable')
plt.ylabel('Y lable')
plt.title('TITLE')
```

```python
# multiple plot 

plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# object oriented methode

fig=plt.figure()

axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)

axes.set_xlabel('x label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])    #add_axes([left,bottom,width,hight])
axes1.set_title('title 1')
axes1.plot(x,y)
axes1.set_xlabel('X label')
axes1.set_ylabel('Y label')


axes2=fig.add_axes([0.2,0.39,0.32,0.25]) 
axes2.plot(y,x)
axes2.set_title('title 2')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=4)
plt.tight_layout()

for current_ex in axes:
    current_ex.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[1].plot(x,y)
axes[1].set_title('second plot')

axes[0].set_title('first plot')
axes[0].plot(y,x)
```

# Figure size and DPI

```python
fig=plt.figure(figsize=(6,2))   # figsize(width,hight)

ax=fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=3,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout
```

#  customization colour

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,color='black',lw=4,alpha=0.9,ls='-',marker='o',markersize=10,
       markerfacecolor='yellow',markeredgewidth=3)  # RGB Hex code  # lw== linewidth  ls==linestyle 
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,color='black',lw=2,ls='--',marker='o')
```
